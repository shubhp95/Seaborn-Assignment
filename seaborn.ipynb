```python
#Ans1:
#a.Scatter Plot: Visualizes relationships and trends between two variables.
#b.Histogram: Analyzes the distribution of a single variable.
#c.Box Plot: Summarizes distribution, detects outliers, and compares groups.
#d.Heatmap: Displays data density and correlations in matrix form.
#e.Pair Plot: Examines relationships between multiple variables and their distributions.
```

```python
#Ans2:
import seaborn as sns
import matplotlib.pyplot as plt
fmri = sns.load_dataset('fmri')
plt.figure(figsize=(12, 6))
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region', markers=True)

plt.title('Line Plot of Signal over Timepoint by Event and Region')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.grid(True)
plt.show()
```

```python
#Ans3:
titanic = sns.load_dataset('titanic')

plt.figure(figsize=(12, 6))

# Box plot for 'age' by 'pclass'
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()

plt.show()
```

```python
#Ans4:
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=30)

plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.grid(True) 
plt.show()
```

```python
#Ans5:
iris = sns.load_dataset('iris')

sns.pairplot(iris, hue='species')

plt.suptitle('Pair Plot of Iris Dataset', y=1.02)  # Adjust title position

plt.show()
```

```python
flights = sns.load_dataset('flights')
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)

plt.title('Heatmap of Number of Passengers by Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```
